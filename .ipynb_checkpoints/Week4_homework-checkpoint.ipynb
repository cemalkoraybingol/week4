```python
import bs4
import pandas as pd
import requests
import csv
```

```python
#YOK mimarlik programi bulunan uni listersi sayfasi
response = requests.get('https://yokatlas.yok.gov.tr/lisans-bolum.php?b=10155')
print(response.text) # Print the output
```

```python
#HTML Parse
soup = bs4.BeautifulSoup(response.text, "html.parser")
print(soup.prettify()) # Print the output using the 'prettify' function
```

```python
# Access the title element
soup.title
```

```python
# Access the content of the title element
soup.title.string
```

```python
# Butun okullarin sayfalarinin linklerini topluyorum
soup.find_all('h4' , attrs={'href' : 'link'})
schoollinks = []
for link in soup.find_all('a'):
    if link.has_attr('href'):
        #print (link.attrs['href'])
        if 'lisans.php' in link.attrs['href']:
            print(link.attrs['href'])
            schoollinks.append(link.attrs['href'])
print(type(schoollinks[0]))
#schoollinks
```

```python
#ilk linki acip icinde
responseOkul = requests.get('https://yokatlas.yok.gov.tr/'+schoollinks[0])
print(responseOkul.text) # Print the output
```

```python
data = []
responseOkul = requests.get('https://yokatlas.yok.gov.tr/content/lisans-dynamic/1000_1.php?y='+schoollinks[0][13:])
okul = bs4.BeautifulSoup(responseOkul.text, "html.parser")
#print(okul.prettify()) # Print the output using the 'prettify' func
# Access the title element
title = okul.find_all(attrs={'class' : "thb text-left"})
for i in range(len(title)):
    title[i] = title[i].string

data.append(title)
```

```python
#bu srtring yapilip request gonderilecek
for num in schoollinks:
    responseOkul = requests.get('https://yokatlas.yok.gov.tr/content/lisans-dynamic/1000_1.php?y='+num[13:])
    okul = bs4.BeautifulSoup(responseOkul.text, "html.parser")
    result = okul.find_all(attrs={'class':"text-center vert-align"})
    for i in range(len(result)):
        result[i] = result[i].string
    data.append(result)
```

```python
data
```

```python
yerlesen = 0
print(len(data))
for i in range(1,len(data)):
    if data[i][9] == None:
        data[i][9] = '0'
    yerlesen+=int(data[i][9])

    
print(yerlesen)
```

```python
f = open('yok_mimarlik_data.csv','a') # open new file, make sure path to your data file is correct

p = 0 # initial place in array
l = len(data)-1 # length of array minus one


f.write("County, State, FIPS Code, Total Pop, Median Income ($), No. of Housing Units, Median Home Value ($), No. of Owner Occupied Housing Units, No. of Owner Occ. Housing Units with Debt, No. of Owner Occ. Housing Units without Debt\n") #write headers


while p < l: # while place is less than length
    f.write(data[p].string + ", ") # write county and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write FIPS and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write Total Pop and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write Median Income and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write No. of Housing Units and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write Median Home Value and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write No. of Owner Occupied Housing Units and add comma
    p = p + 1 # increment
    f.write(data[p].string + ", ") # write No. of Owner Occ. Housing Units with Debt and add comma
    p = p + 1 # increment
    f.write(data[p].string + "\n") # write No. of Owner Occ. Housing Units without Debt and line break
    p = p + 1 # increment

    
f.close() # close file
```
